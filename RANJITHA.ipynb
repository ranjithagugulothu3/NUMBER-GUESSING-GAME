```python
import random
print("ğŸ¯ Welcome to the Number Guessing Game!")
print("I am thinking of a number between 1 and 100.")
secret_number = random.randint(1, 100)
max_attempts = 6
attempts = 0
score = 100
while attempts < max_attempts:
    guess = int(input("\nEnter your guess: "))
    attempts += 1

    if guess == secret_number:
        print("ğŸ‰ Congratulations! You guessed the correct number.")
        break
    elif guess < secret_number:
        print("ğŸ“‰ Too low! Try a higher number.")
    else:
        print("ğŸ“ˆ Too high! Try a lower number.")
    score -= 10
    print(f"Remaining attempts: {max_attempts - attempts}")
if attempts == max_attempts and guess != secret_number:
    print("\nâŒ Game Over!")
    print(f"The correct number was: {secret_number}")
print(f"\nğŸ† Your final score is: {score}")
```
